```python
import optuna
import matplotlib.pyplot as plt
import numpy as np
```

# EQUATION SOLVING
In this section we different equations will be solved. Due to the stochaistic nature of optuna, each equation will be solved 5 times and the average of the scores will be calculated.

```python
iterations = 5
optimization_time = 12
optimization_direction = 'minimize'
```

## 1. EQUATION

```python
def objective(trial):
    x = trial.suggest_float('x', -100.0, 100.0)
    y = trial.suggest_float('y', -100.0, 100.0)
    z = trial.suggest_float('z', -100.0, 100.0)
    k = trial.suggest_float('k', -100.0, 100.0)

    return (x**2 - 4*y**3 / z**4) * k**3

best_scores = list()
for iteration in range(iterations):
    print(f'ITERATION {iteration + 1}')
    study = optuna.create_study(direction=optimization_direction)
    study.optimize(objective, timeout=optimization_time)
    
    best_score = study.best_value
    best_scores.append(best_score)
```

```python
print(f'BEST SCORES: {best_scores}')
print(f'MEAN BEST SCORE: {np.mean(best_scores)}')
print(f'MAX SCORE: {np.max(best_scores)}')
print(f'MIN SCORE: {np.min(best_scores)}')
print(f'STD SCORES: {np.std(best_scores)}')
```

## 2. EQUATION

```python
def objective(trial):
    x = trial.suggest_float('x', -100.0, 100.0)
    y = trial.suggest_float('y', -100.0, 100.0)
    z = trial.suggest_float('z', -100.0, 100.0)
    k = trial.suggest_float('k', -100.0, 100.0)

    return np.cos(x) * np.cos(y) * np.cos(z) * np.cos(k)

best_scores = list()
for iteration in range(iterations):
    print(f'ITERATION {iteration + 1}')
    study = optuna.create_study(direction=optimization_direction)
    study.optimize(objective, timeout=optimization_time)
    
    best_score = study.best_value
    best_scores.append(best_score)
```

```python
print(f'BEST SCORES: {best_scores}')
print(f'MEAN BEST SCORE: {np.mean(best_scores)}')
print(f'MAX SCORE: {np.max(best_scores)}')
print(f'MIN SCORE: {np.min(best_scores)}')
print(f'STD SCORES: {np.std(best_scores)}')
```
